# Spotify REST API Challenge

__What to listen?__

Create your own playlist based on the related artists and their most popular tracks using the [Spotify REST API](https://developer.spotify.com/)

```python
# Import
import pandas as pd
import requests
```

### Get access!!!

Get your `client_id` and `client_secret` to generate your __token__ access

```python
# Generate token with a POST request

client_id = "32ba556c59b240adba9423ce4c752500" # CLIENT ID
client_secret = "d587e7c304f9484f9022742fba0767da" # CLIENT SECRET
auth_url = 'https://accounts.spotify.com/api/token'
```

```python
auth_response = requests.post(auth_url, {'grant_type': 'client_credentials',
                                         'client_id': client_id,
                                         'client_secret': client_secret}).json()
```

```python
access_token = auth_response['access_token']
auth_response
```

### Set your main variables!!!

Set the `base_uri` (i.e.: end-point), parameters and `headers` for your __GET__ operations

```python
# Base end-point construction

base_url = 'https://api.spotify.com/v1/'
resource = 'artists/'

header_info = {'Authorization': 'Bearer {token}'.format(token=access_token)}
```

### Create your new playlist!!!

```python
# Here you can complete your endpoint URI

full_endpoint = base_url + resource + "0TnOYISbd1XYRBk9myaseg/related-artists"
full_endpoint
```

```python
# Start building your playlist!!!

r = requests.get(full_endpoint, headers=header_info)
r.status_code
```

```python
r_json=r.json()
type(r_json)
```

```python
r_list= r.json()["artists"]
r_list
```

```python
for i in range(len(r_list)):
    print(r_list[i] ["name"])
```

```python
df_relate_artists =pd.DataFrame(r_list)
df_relate_artists
```

```python
top_tracks=base_url + resource + "0TnOYISbd1XYRBk9myaseg/top-tracks"
top_tracks
```

```python
t= requests.get(top_tracks, headers=header_info)
t.status_code
print(t)

```

```python
t_json=t.json()
type(t_json)
```

```python
t_json.keys()
```

```python
t_track = t.json()["tracks"]
t_track
```
